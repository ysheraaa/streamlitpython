```python
import pandas as pd
```

```python
penguin = pd.read_csv('penguin.csv')
```

```python
data = penguin.copy()
```

```python
target = 'jenis'
```

```python
encode = ['gender', 'pulau']
```

```python
data
```

```python
for col in encode: 
    dummy = pd.get_dummies(data[col], prefix=col)
    data = pd.concat([data,dummy], axis=1)
    del data[col]
```

```python
data
```

```python
target_mapper = {'Adelie' : 0, 'Chinstrap' : 1, 'Gentoo' : 2}
```

```python
def target_encode(val):
    return target_mapper[val]
```

```python
data['jenis'] = data['jenis'].apply(target_encode)
```

```python
data
```

```python
X = data.drop('jenis', axis=1)
Y = data['jenis']
```

```python
X
```

```python
Y
```

```python
# Model Naive Bayes Classification
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X, Y)
```

```python
# Untuk Menyimpan Model Naive Bayes Classification NBC
import pickle 
pickle.dump(model, open('modelNBC_penguin.pkl', 'wb'))
#wb write binary untuk menyimpan, kalau rp untuk membaca 
```
